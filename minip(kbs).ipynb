1.	How many houses are in the range of prices from 0-25lakhs, 26-50lakhs, 51-75lakhs, 76-100lakhs and >100lakhs. Visualize it using a line chart also.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = '/content/Housing.csv'
data=pd.read_csv(file_path)
plt.figure(figsize=(10,5))
price=[0,2500000,5000000,7500000,10000000,float('inf')]
labels=["0-25L", "26-50L", "51-75L", "76-100L", ">100L"]
data['Price Category']=pd.cut(data['price'], bins=price, labels=labels)
counts=data['Price Category'].value_counts().sort_index()
plt.figure(figsize=(10,5))
plt.plot(counts.index,counts.values,marker='o',linestyle='-')
plt.xlabel("Price Range of houses")
plt.ylabel("Number of Houses")
plt.title("Distribution of houses by price range")
plt.grid()
plt.legend('Price')
plt.show()
```

2.	Find average house prices for houses having AC and nonAC. Plot the relationship using a barchart.

```python
prices=data.groupby('airconditioning')['price'].mean()
plt.figure(figsize=(6,5))
prices.plot(kind='bar',color=['blue', 'green'])
plt.xlabel("AC")
plt.ylabel("No.of houses")
plt.title("Houses Prices Based on Air Conditioning")
plt.grid()
plt.show()
```

3.	Simulate the relationship between parking and house price.

```python
plt.figure(figsize=(6,5))
plt.bar(data['parking'],data['price'])
labels=['0','1','2','3']
plt.xticks(range(len(labels)), labels)
plt.xlabel("Parking slot")
plt.ylabel("Number of houses")
plt.title("Parking slot based on number of houses")
plt.grid()
plt.show()
```

4.	What is price gap between the houses having <5000sqft & no prefarea , and >5000sqft & there is prefarea?

```python
a1=data[(data['area']<5000)&(data['prefarea']=="no")]['price'].sum()
a2=data[(data['area']>5000)&(data['prefarea']=="yes")]['price'].sum()
print('Total of group1:',a1)
print('Total of group2:',a2)
a3=data[(data['area']<5000)&(data['prefarea']=='no')]['price'].mean()
a4=data[(data['area']>5000)&(data['prefarea']=='yes')]['price'].mean()
print('Average of a3:',a3)
print('Average of a4:',a4)
price_gap=a4-a3
print(f"Price gap between <5000 sqft & no prefarea vs >5000 sqft & prefarea: ",price_gap)
```
